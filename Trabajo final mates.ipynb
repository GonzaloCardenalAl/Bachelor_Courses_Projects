```python
a, c, b, x, y, z = sp.symbols('a c b x y z')
```

```python
f_libro = c *(-y + x - (x**3)/3 + z)
g_libro = -((x - a + b*y)/ c)
```

```python
#pto_c = sp.simplify(sp.solve([f_libro, g_libro], [x, y]))    
#pto_c
```

```python
Jac_libro = sp.Matrix([[f_libro.diff(x), f_libro.diff(y)], [g_libro.diff(x), g_libro.diff(y)]])
sp.simplify(Jac_libro)
```

Las nullclines para este sistema son las siguientes:

La nullcline y tiene la forma:

$\frac{dy}{dt} = 0$ -> $y = \frac{1}{b}x + \frac{a}{b}$

Esta nullcline dibuja una linea recta con pendiente $\frac{1}{b}$ y que corta el eje y en el punto $(0,\frac{a}{b})$

La nullcline x tiene la forma:

$\frac{dx}{dt} = 0$ -> $y = x - \frac{1}{3}x^3 + z$

Como vemos en ambas nullclinas, el parámetro $c$ no aparece en ninguna de las 2, por lo que el valor de $c$ no afecta a la forma de las curvas. Veremos que, aunque no afecte a la forma de la mismas, sí que lo hace en la velocidad de movimiento en los planos de fase.

La pendiente de la curva cúbica  es la siguiente:
$$M(x) = \frac{dy}{dx} = 1 - x^2$$

Por tanto esta ecuación cúbica tiene extremos (también llamados puntos críticos) cuando:
$$ M(x) = 0$$
$$ 1-x^2 = 0 $$
$$ x= ±1 $$

También podemos observar que la segunda derivada $\frac{d^2y}{d^2x}=-2x$, por lo que es positiva cuando $x=-1$ y corresponde a un mínimo y es negativa cuando $x=1$ y, por lo tanto, se corresponde con un máximo. Este resultado no se ve afectado por el valor de ninguno de los parámetros del modelo.
