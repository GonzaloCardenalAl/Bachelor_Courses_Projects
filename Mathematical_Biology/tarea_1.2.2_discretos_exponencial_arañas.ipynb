### Cobweb maps
En estos cobweb maps se está representando pn (coordenada x) frente a pn+1 (coornada y), con lo que podemos observar los distintos comportamientos de los modelos (tanto cerrado como abierto) en función de la constante rho.
Además añadimos la línea de la primera iteración que representa la población incial de la que partimos.

### 1.1. Crecimiento exponencial  

Esto corresponde al caso $1<\rho$

```python
    import matplotlib.pyplot as plt
    from numpy import cos, linspace
    
    def f(x):
        rho = 1.2
        x = rho*x
        return x
    
    def cobweb(f, p0, N, a=0, b=40):
        #definición de la función y sus parámetros
        # plot the function being iterated
        t = linspace(a, b, N)
        plt.plot(t, f(t), 'k')

        # plot the dotted line y = x
        plt.plot(t, t, "k:")
        
        # Pintamos la primera iteración
        plt.plot([p0],[0], 'bo')
        plt.plot([p0,p0], [0, f(p0)], 'b', linewidth=1)
    
    
        # plot the iterates
        x, y = p0, f(p0)
        for _ in range(N):
            fy = f(y)        
            plt.plot([x, y], [y,  y], 'b', linewidth=1)
            plt.plot([y, y], [y, fy], 'b', linewidth=1)
            x, y = y, fy
             
        plt.show()
        plt.close()
        
p0=1
N=20       

cobweb(f, p0, N)
```

Para $\rho>1$, la población tiene crecimiento exponencial e infinito.

### 1.3. Oscilaciones con crecimiento exponencial  (sin sentido biológico)

### 2. Modelos abiertos con inmigración

En estos modelos, la solución de equilibrio es k= I/(1-$\rho$)

### 2.1. Crecimiento exponencial  con migraciones
Esto corresponde al caso $1<\rho$

```python
    import matplotlib.pyplot as plt
    from numpy import cos, linspace
    
    def f(x):
        I = 1
        rho = 1.2
        x = rho*x + I
        return x
    
    def cobweb(f, p0, N, a=0, b=250):
        #definición de la función y sus parámetros
        # plot the function being iterated
        t = linspace(a, b, N)
        plt.plot(t, f(t), 'k')

        # plot the dotted line y = x
        plt.plot(t, t, "k:")
        
        # Pintamos la primera iteración
        plt.plot([p0],[0], 'bo')
        plt.plot([p0,p0], [0, f(p0)], 'b', linewidth=1)
    
        # plot the iterates
        x, y = p0, f(p0)
        for _ in range(N):
            fy = f(y)        
            plt.plot([x, y], [y,  y], 'b', linewidth=1)
            plt.plot([y, y], [y, fy], 'b', linewidth=1)
            x, y = y, fy
             
        plt.show()
        plt.close()
        
p0=1
N=20       

cobweb(f, p0, N)
```

Para $1>\rho$,  La población tiende a infinito con mayor crecimiento debido a las inmigraciones.
