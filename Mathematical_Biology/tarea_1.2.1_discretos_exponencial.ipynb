# Tarea 1.2.1: Comportamientos de modelos discretos simples de una población



1. Representar todos los posibles comportamientos de los modelos discretos de crecimineto y decrecimiento exponencial en el plano $(n, p_n)$

    Dado que estos modelos son de la forma $p_{n+1} = \rho p_n$ la población en cada instante $n$ está dada por 

    $$
    p_n = p_0 \rho^n
    $$

    En esta expresión $p_0$ es la población en el instante inicial.



2.  Representar todos los posibles comportamientos de los modelos discretos de crecimineto y decrecimiento exponencial con migración constante en el plano $(n, p_n)$
    
    Dado que estos modelos son de la forma $p_{n+1} = \rho p_n + I$ la población en cada instante $n$ está dada por 

    $$
    p_n = \left(p_0 - \frac{I}{1-\rho}\right) \rho^n + \frac{I}{1-\rho}
    $$
    
    En esta expresión $I$ es la inmigración.

Primero cargamos el módulo que vamos a usar

```python
import matplotlib.pyplot as plt
```

### 1.1. Crecimiento exponencial  

Esto corresponde al caso $1<\rho$

```python
ro = 1.2
p0 = 1
tiempo = [t for t in range(10)]
poblacion = [p0*ro**n for n in range(10)]
#plt.scatter(tiempo, poblacion, color = 'red')
plt.plot(tiempo, poblacion, 'o-', color = 'red', linewidth = 0.5)
plt.show()
```

Para $\rho>1$, la población tiene crecimiento exponencial e infinito.

## 1.2. Extinción exponencial  

Esto corresponde al caso $0<\rho<1$

```python
ro = 0.5
p0 = 1
tiempo = [t for t in range(10)]
poblacion = [p0*ro**n for n in range(10)]
#plt.scatter(tiempo, poblacion, color = 'red')
plt.plot(tiempo, poblacion, 'o-', color = 'red', linewidth = 0.5)
plt.show()
```

Para $0<\rho<1$, la población va a decrecer hasta la extinción.

### 1.3. Oscilaciones con crecimiento exponencial  (sin sentido biológico)

Esto corresponde al caso $\rho<-1$

```python
ro = -1.2
p0 = 1
tiempo = [t for t in range(10)]
poblacion = [p0*ro**n for n in range(10)]
#plt.scatter(tiempo, poblacion, color = 'red')
plt.plot(tiempo, poblacion, 'o-', color = 'red', linewidth = 0.5)
plt.show()
```

Para $\rho<-1$, el modelo no tiene sentido biológico. La población tiene oscilaciones divergentes que tienden a infinito.

### 1.4. Oscilaciones con decrecimiento exponencial  (sin sentido biológico)

Esto corresponde al caso $-1 < \rho < 0$

```python
ro = -0.8
p0 = 1
tiempo = [t for t in range(10)]
poblacion = [p0*ro**n for n in range(10)]
#plt.scatter(tiempo, poblacion, color = 'red')
plt.plot(tiempo, poblacion, 'o-', color = 'red', linewidth = 0.5)
plt.show()
```

Para $-1<\rho<0$, el modelo no tiene sentido biológico. La población tiene oscilaciones convergentes a 0.

### 2.1. Crecimiento exponencial  con migraciones

Esto corresponde al caso $1<\rho$

```python
ro = 1.1
p0_1 = - 820
p0_2 = 40
I = 50

K = I/(1 - ro)
c1 = (p0_1 - K)
c2 = (p0_2 - K)

tiempo = [t for t in range(15)]
poblacion1 = [c1*ro**n + K for n in tiempo]
poblacion2 = [c2*ro**n + K for n in tiempo]

plt.scatter(tiempo, poblacion1, label = '$p_0 =$' +  str(p0_1))
plt.plot(tiempo, [K for n in tiempo], '-o', color = 'red', label = 'Solución de equilibrio, K=' + str(round(K)))
plt.scatter(tiempo, poblacion2, label = '$p_0 =$' + str(p0_2))
plt.legend()
plt.show()
```

Para $1<\rho$,  si la población inicial está por encima de K la población diverge a infinito, y si la población inicial está por debajo de K (sin sentido biológico) la población diverge hacia -infinito. La solución de equilibrio siempre da negativo por lo que no tiene sentido biológico.

## 2.2. Decrecimiento exponencial con migraciones

Esto corresponde al caso $0<\rho<1$

```python
ro = 0.7
p0_1 = - 820
p0_2 = 300
I = 50

K = I/(1 - ro)
c1 = (p0_1 - K)
c2 = (p0_2 - K)

tiempo = [t for t in range(15)]
poblacion1 = [c1*ro**n + K for n in tiempo]
poblacion2 = [c2*ro**n + K for n in tiempo]

plt.scatter(tiempo, poblacion1, label = '$p_0 =$' +  str(p0_1))
plt.plot(tiempo, [K for n in tiempo], '-o', color = 'red', label = 'Solución de equilibrio, K=' + str(round(K)))
plt.scatter(tiempo, poblacion2, label = '$p_0 =$' + str(p0_2))
plt.legend()
plt.show()
```

Para $0<\rho<1$, la población converge a la solución de equilibrio K debido a las inmigraciones.

### 2.3. Oscilaciones con crecimiento exponencial y migraciones

Esto corresponde al caso $\rho<-1$

```python
ro = -1.2
p0_1 = - 820
p0_2 = 350
I = 50

K = I/(1 - ro)
c1 = (p0_1 - K)
c2 = (p0_2 - K)

tiempo = [t for t in range(20)]
poblacion1 = [c1*ro**n + K for n in tiempo]
poblacion2 = [c2*ro**n + K for n in tiempo]

plt.scatter(tiempo, poblacion1, label = '$p_0 =$' +  str(p0_1))
plt.plot(tiempo, [K for n in tiempo], '-o', color = 'red', label = 'Solución de equilibrio, K=' + str(round(K)))
plt.scatter(tiempo, poblacion2, label = '$p_0 =$' + str(p0_2))
plt.legend()
plt.show()
```

Para $\rho<-1$, el sistema tiene oscilaciones divergentes sin sentido biológico. Si el número de generación es impar, la población dará negativo y si es par, positivo.

### 2.4. Oscilaciones con decrecimiento exponencial y migraciones

Esto corresponde al caso $-1 < \rho < 0$

```python
ro = -0.8
p0_1 = - 20
p0_2 = 40
I = 50

K = I/(1 - ro)
c1 = (p0_1 - K)
c2 = (p0_2 - K)

tiempo = [t for t in range(15)]
poblacion1 = [c1*ro**n + K for n in tiempo]
poblacion2 = [c2*ro**n + K for n in tiempo]

plt.scatter(tiempo, poblacion1, label = '$p_0 =$' +  str(p0_1))
plt.plot(tiempo, [K for n in tiempo], '-o', color = 'red', label = 'Solución de equilibrio, K=' + str(round(K)))
plt.scatter(tiempo, poblacion2, label = '$p_0 =$' + str(p0_2))
plt.legend()
plt.show()
```

Para $-1 < \rho < 0$, la población tiene oscilaciones convergentes sin sentido biológico. Si el número de generación es impar, la población dará negativo y si es par, positivo.
